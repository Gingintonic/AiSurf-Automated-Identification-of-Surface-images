# Calculate SIFT-descriptors at detected keypoints

```python
# merge dark and light blobs
kps = np.concatenate((blobs_log,blobs_log_negative))
# prepare everthing for opencv
# skimage somehow thinks switching x and y is funny
kps_cv = [cv.KeyPoint(kp[1],kp[0],kp[2]) for kp in kps]
for kp in kps_cv:
    kp.angle = 0
img_cv = cv.imread('./experiments/small SrTiO3_1244.png')
img_cv_gray = cv.cvtColor(img_cv, cv.COLOR_BGR2GRAY)
```

```python
# parameters for sift descriptor
cThr = 0.003
sigma = 4
nOctLayers = 3

sift = cv.xfeatures2d.SIFT_create(contrastThreshold = cThr,
                                  sigma = sigma, 
                                  nOctaveLayers = nOctLayers)
# compute descriptors
_, desc = sift.compute(img_cv_gray,kps_cv)
```

```python
from sklearn.cluster import AgglomerativeClustering as Agg_Cl
from sklearn.metrics import silhouette_score

def find_best_clustering(descriptors, 
                         span = range(2,30), 
                         sklearn_clustering = Agg_Cl, 
                         **kwargs):
    '''
    Finds the optimal number of clusters of the given descriptors
    wrt to the silhouette score
    
    Parameters:
    -----------
    descriptors - descriptors of some features, 
                  an nxd array for n features with d descriptors 
    span        - an iterable range of values, 
    **kwargs    - additional possible arguments for AgglomerativeClustering
    
    Result:
    -------
    best_labels - for each entry in descriptor returns a 
                  label specifying the described features cluster
    '''
    best_score = -1
    for N in span:
        clustering = sklearn_clustering(n_clusters = N,
                                        **kwargs).fit(descriptors)
        score = silhouette_score(descriptors,
                                 clustering.labels_)
        if score > best_score:
            best_labels = clustering.labels_
            best_score = score
            best_N = N
    print(
        "Max score found @ {} clusters:".format(best_N),
        "\nsil_score = {}".format(best_score))
    return best_labels
```

```python
# number of clusters controls how specific
# differentiation between local environments
# should be
# e.g. 2 --> only light and dark
#      9 --> adjacent environment comes into play
labels = find_best_clustering(desc, span=range(2,10))
```

```python
plt.imshow(img_cv_gray,cmap='gray')
for i in set(labels):
    plt.scatter(kps[labels==i,1],kps[labels==i,0],s=kps[labels==i,2])
```

# Extract lattice

```python
#cluster label with maximum number of keypoints:
l_max = np.argmax(np.bincount(labels))
#l_max = 6

plt.imshow(img_cv_gray,cmap='gray')
plt.scatter(kps[labels==l_max,1],
            kps[labels==l_max,0],
            s=1,
            color='red')
```

```python
# get coordinates of kps and of most frequent cluster
x = np.array([kp.pt for kp in kps_cv])
x_ofMaxCluster = x[labels==l_max]
k1 = 6
kNN = utilities.kNearestNeighbours(x_ofMaxCluster,k1)
kNN_pos_x = np.where(np.repeat(kNN[:,0,np.newaxis]<0, 2, axis=1), -kNN, kNN)

#plt.figure(figsize=(4,4),dpi=160)
plt.axis("equal")
plt.gca().invert_yaxis()
plt.grid()
plt.scatter(kNN[:,0],kNN[:,1],s=1,label='kNN')
plt.scatter(kNN_pos_x[:,0],kNN_pos_x[:,1],s=1,label='kNN pos x')
plt.xlabel("x [px]")
plt.ylabel("y [px]")
plt.legend()
plt.show()
```

```python
min_cluster_size = 5

print("Clustering NNV's of maximally represented feature class:")
cluster_span_kNN = range(6,30)
kNN_labels = find_best_clustering(kNN_pos_x,span=cluster_span_kNN)

#the potential lattice vectors are the average vectors of each cluster:
average_peaks = [np.mean(kNN_pos_x[kNN_labels==l],axis=0) 
                 for l in set(kNN_labels) 
                 if np.sum(labels==l)>min_cluster_size]
average_peaks = np.array(average_peaks)
```

```python
plt.axis("equal")
plt.gca().invert_yaxis()
plt.grid()
plt.scatter(kNN_pos_x[:,0],kNN_pos_x[:,1],s=1,label='kNN pos x')
for peak in average_peaks:
    plt.scatter(peak[0],peak[1])
plt.xlabel("x [px]")
plt.ylabel("y [px]")
plt.legend()
plt.show()
```

```python
n,m = average_peaks.shape
distance_matrix = np.zeros((n,n,m))

for i,peak_1 in enumerate(average_peaks):
    for j,peak_2 in enumerate(average_peaks):
        distance_matrix[i,j] = peak_1 - peak_2
```

```python
distance_vectors = []
for i in range(0,n):
    for j in range(i+1,n):
        distance_vectors.append(distance_matrix[i,j])
distance_vectors = np.array(distance_vectors)
distance_vectors_length = np.linalg.norm(distance_vectors,axis=1)
#distance_vectors = distance_vectors[np.argsort(distance_vectors_length)]

potential_lattice_vectors = []
        
```

```python
min_tresh = min(np.linalg.norm(distance_vectors,axis=1))*1.10
```

```python
candidates = distance_vectors[distance_vectors_length< min_tresh]
candidates
```

```python
candidates_sorted = np.sort(np.where((candidates[:,0]>0).T,candidates.T,candidates.T*-1).T,axis=0)
candidates_sorted
```

```python
partition = np.argmax(np.diff(candidates_sorted,axis=0),axis=0)
```

```python
a = np.mean(candidates_sorted[:partition[0]],axis=0)
b = np.mean(candidates_sorted[partition[0]+1:],axis=0)
print(a,b)
```
